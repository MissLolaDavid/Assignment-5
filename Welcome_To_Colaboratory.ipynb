```
print("Enter [P] to see available options below")


def the_options():
    print("Options: ")
    print("[C] Convert from Celsius")
    print("[F] Convert from Fahrenheit")
    print("[Mi] Convert from Miles")
    print("[KM] Convert from Kilometers")
    print("[CM_M] Convert from Centimeters")
    print("[M] Convert from Meters")
    print("[YD] Convert from Yard")
    print("[M2] Convert from Meters")
    print("[In] Convert from Inches")
    print("[CM] Convert from Centimeters")
    print("[Q] Quit")


converting = True
while converting:
    user_input = str.upper(input("Options: "))
    if user_input == "P":
        the_options()
    elif user_input == "C":
        def celsius():
            cel = float(input("Enter in Celsius Unit: "))
            calc_cel = (cel * 1.8) + 32
            print(str(calc_cel) + " Fahrenheit")


        celsius()
    elif user_input == "F":
        def fahrenheit():
            f = float(input("Enter in Fahrenheit Unit: "))
            calc_f = (f - 32) * 1.8
            print(str(calc_f) + " Celsius")


        fahrenheit()
    elif user_input == "Mi":
        def miles():
            mi = float(input("Enter in Miles Unit: "))
            calc_mi = float(mi * 1.609)
            print(str(calc_mi) + " Kilometers")


        miles()
    elif user_input == "KM":
        def kilometers():
            km = float(input("Enter in Kilometers Unit: "))
            calc_km = float(km / 1.609)
            print(str(calc_km) + " Miles")


        kilometers()
    elif user_input == "CM_M":
        def centimeters():
            CM_M = float(input("Enter in Centimeters Unit: "))
            calc_cm_m = float(CM_M / 100)
            print(str(calc_cm_m) + " Meters")


        centimeters()
    elif user_input == "M":
        def meters():
            M = float(input("Enter in Meters Unit: "))
            calc_m = float(M * 100)
            print(str(calc_m) + " Centimeters")


        meters()
    elif user_input == "YD":
        def yard():
            YD = float(input("Enter in Yard Unit: "))
            calc_yd = float(YD / 1.094)
            print(str(calc_yd) + " Meters")


        yard()
    elif user_input == "M2":
        def meter():
            M2 = float(input("Enter in Meter Unit: "))
            calc_m2 = float(M2 * 1.094)
            print(str(calc_m2) + " Yards")


        meter()
    elif user_input == "In":
        def inches():
            In = float(input("Enter in Inches Unit: "))
            calc_in = float(In * 2.54)
            print(str(calc_in) + " Centimeters")


        inches()
    elif user_input == "CM":
        def centimeters():
            CM = float(input("Enter in Centimeter Unit: "))
            calc_cm = float(CM / 2.54)
            print(str(calc_cm) + " Inches")


        centimeters()
    elif user_input == "Q":
        converting = False
        break
    else:
        converting = False
```
